```python
import json

def get_metrics_dict_rrt(file_path): #file_path = 'RESULTS/SCENE1/4dof/a-star/scene1_data.json'

    with open(file_path, 'r') as file:
        data = json.load(file)

    # Initialize a dictionary to store lists for each metric
    metrics_dict = {
        "number_of_IK_results": [],
        "number_of_collision_check": [],
        "number_of_collision_check_in_IK": [],
        "number_of_denied_nodes_goal": [],
        "number_of_denied_nodes_random": [],
        "number_of_goal_expanding_nodes": [],
        "number_of_nn_check": [],
        "number_of_random_nodes": [],
        "opened_nodes": [],
        "reached_goal": [],
        "time_of_IK_results": [],
        "time_of_collision_check": [],
        "time_of_collision_check_in_IK": [],
        "time_of_nn_check": [],
        "timeout": [],
        "total_duration": [],
    }

    for optimization_run in data:
        for key in metrics_dict:
            metrics_dict[key].append(float(optimization_run[key]))
    return metrics_dict



def get_metrics_dict_astar(file_path): #file_path = 'RESULTS/SCENE1/4dof/a-star/scene1_data.json'

    with open(file_path, 'r') as file:
        data = json.load(file)

    # Initialize a dictionary to store lists for each metric
    metrics_dict = {
        "closed_nodes": [],
        "coll_check_percentage": [],
        "coord_tolerance": [],
        "g_cost": [],
        "g_units": [],
        "opened_nodes": [],
        "total_time": [],
        "turn_number": []
    }

    # Iterate through each optimization run
    for optimization_run in data:
        for key in metrics_dict:
            if optimization_run[key]!=0:
                metrics_dict[key].append(optimization_run[key])
    # for optimization_run in data:
    #     # Append values to respective lists in the dictionary
    #     metrics_dict["closed_nodes"].append(optimization_run["closed_nodes"])
    #     metrics_dict["coll_check_percentage"].append(optimization_run["coll_check_percentage"])
    #     metrics_dict["coord_tolerance"].append(optimization_run["coord_tolerance"])
    #     metrics_dict["g_cost"].append(optimization_run["g_cost"])
    #     metrics_dict["g_units"].append(optimization_run["g_units"])
    #     metrics_dict["opened_nodes"].append(optimization_run["opened_nodes"])
    #     metrics_dict["total_time"].append(optimization_run["total_time"])
    #     metrics_dict["turn_number"].append(optimization_run["turn_number"])

    # Now you have a dictionary with metric names as keys and lists of corresponding values
    return metrics_dict
```

```python
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence_level=0.95):
    degrees_of_freedom = len(data) - 1  # for t-distribution

    # Calculate mean and standard error
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # ddof=1 for sample standard deviation
    standard_error = sample_std / np.sqrt(len(data))

    # Calculate critical value from t-distribution
    t_critical = t.ppf((1 + confidence_level) / 2, degrees_of_freedom)

    # Calculate confidence interval
    margin_of_error = t_critical * standard_error
    return margin_of_error
```
